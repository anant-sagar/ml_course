```python
from sklearn.impute import SimpleImputer
```

```python
sim_imp = SimpleImputer(strategy="mean")
```

```python
sim_imp.fit_transform(df[["Age"]])
```

```python
def salary_clean(salary_1):
  if isinstance(salary_1,str):
    if "$" in salary_1:
      salary_1= salary_1.replace("$","")
    if "," in salary_1:
      salary_1= salary_1.replace(",","")
    return int(salary_1)
```

```python
df["salary"] = df["salary"].apply(salary_clean)
```

```python
sim_imp.fit_transform(df[["salary"]])
```

```python
cols = ["Age","salary"]
```

```python
df[cols]=sim_imp.fit_transform(df[cols])
```
